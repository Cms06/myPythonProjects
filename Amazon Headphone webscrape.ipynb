```python
from bs4 import BeautifulSoup
import requests


headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.5",
    "Referer": "https://www.amazon.com/",
    "DNT": "1",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1",
    "TE": "Trailers",
    "Pragma": "no-cache",
    "Cache-Control": "no-cache"
}


output = []

i = 1
while True:
    
    base_url = f'https://www.amazon.com/s?i=electronics-intl-ship&bbn=16225009011&rh=n%3A16225009011%2Cn%3A172541&page={i}&ref=sr_pg_399'

    #request and parse
    page = requests.get(base_url, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    #grab the products list per page
    products = soup.findAll('div', {'data-component-type': "s-search-result"})
    products = products[3:-3]

    #identifier for the last page
    next_button = soup.find('span', {'class': 's-pagination-item s-pagination-next s-pagination-disabled',
                              'aria-disabled': "true"})

    if next_button:
        print('Done!')
        break 
    # looping on the list of products per page and grab the details that we want
    else:
        for product in products:
            try:
                name = product.find(
                    'span', {'class': 'a-size-base-plus a-color-base a-text-normal'}).get_text()
                price = product.find(
                    'span', {'class': 'a-offscreen'}).get_text()
                price = float(price[1:])
                rating = product.find(
                    'span', {'class': 'a-icon-alt'}).get_text()
            # For prices that is empty or no values
            except:
                price = None
            output.append([name, price, rating])
        print(f'page {i} is done!')
        i += 1
        
```
